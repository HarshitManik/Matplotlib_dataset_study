```python
#to calculate median of random integers
y=np.random.randint(0,100,size=(10))
print(y)

y.sort()
print(y)
if len(y) % 2 == 0:  # Even number of elements
    median = (y[len(y)//2 - 1] + y[len(y)//2]) / 2
else:  # Odd number of elements
    median = y[len(y)//2]
print("median:",median)



```
