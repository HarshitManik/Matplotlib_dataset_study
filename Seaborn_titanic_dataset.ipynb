```python
#Web crawling
!pip install seaborn
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
iris=sns.load_dataset  ('iris')
```

```python
iris.head()
sns.distplot(iris['sepal_length'],bins=20,color='red',kde=False)# kde removes the data curve
plt.show()
sns.kdeplot(iris['sepal_length'],shade=True,color='red')
plt.show()
```

```python
sns.scatterplot(x=iris['sepal_length'],y=iris['sepal_width'],hue=iris['species'])# using seaborn for scatter plot hue is used to sert colors for scatter plots
plt.show()
sns.jointplot(x=iris['sepal_length'],y=iris['sepal_width'],hue=iris['species'])# using seaborn for scatter plot hue is used to sert colors for scatter plots
plt.show()
sns.jointplot(x=iris['sepal_length'],y=iris['sepal_width'],kind='hex')# using seaborn for scatter plot hue is used to sert colors for scatter plots
plt.show()
```

```python
# pair plot
sns.pairplot(data=iris,hue='species')
```

```python
# seabor categorical plots
sns.countplot(x='species',data=iris,hue='species')
plt.show()
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.show()
sns.violinplot(x='species',y='sepal_length',data=iris)
plt.show()
sns.barplot(x='species',y='sepal_length',data=iris)
plt.show()
sns.heatmap(iris.corr(),annot=True)
plt.show()
# finding corelation
iris.corr()
```

```python
# VISULAIZE TIPS DATASET
tips=sns.load_dataset('tips')
tips.head()
sns.countplot(x=tips['sex'])
plt.show()
tips['day'].unique()
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
sns.barplot(x='time',y='total_bill',data=tips,hue='sex')
plt.show()
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.scatterplot(y='tip',x='total_bill',data=tips)# the plot shows that if a person has a lesser bil he is likely to give lesser amount of tip
plt.show()
sns.scatterplot(y='tip',x='total_bill',data=tips)
plt.show()# regression line
```

```python
# challeng a titanic dataset
titanic=sns.load_dataset('titanic')
titanic.head()
titanic.shape
sns.countplot(x='sex',data=titanic)
plt.show()
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.show()
sns.violinplot(x='age',y='survived',data=titanic)
```

```python
sns.histplot(titanic, x='age', hue='survived', multiple='stack', kde=True)
plt.title('Distribution of Age by Survival Status')
plt.show()
sns.boxplot(x='class', y='age', hue='survived', data=titanic)
plt.title('Age Distribution by Class and Survival Status')
plt.show()

```
